## 1. What is Deep Learning? <a id='intro'></a>

Deep learning is a subset of machine learning that focuses on neural networks with many layers (hence the term "deep"). These neural networks are designed to automatically and adaptively learn features from data, making them very powerful for tasks such as image and speech recognition, natural language processing, and more.

### Key Concepts:
- **Neural Network**: A network of neurons (or nodes) organized in layers. Each neuron receives input, processes it, and passes the output to the next layer.
- **Layer**: A collection of neurons. The most common types are the input layer, hidden layers, and the output layer.
- **Activation Function**: A function applied to the output of each neuron to introduce non-linearity. Common activation functions include ReLU, Sigmoid, and Tanh.
- **Loss Function**: A function that measures how well the neural network is performing. The goal is to minimize the loss function during training.
- **Optimizer**: An algorithm that adjusts the weights of the neurons to minimize the loss function. Popular optimizers include SGD, Adam, and RMSprop.

## 2. Neural Networks <a id='nn'></a>

A neural network is composed of layers of neurons. The simplest form of a neural network is a feedforward neural network where information moves in one direction, from input to output.

### Example Structure:
1. **Input Layer**: Takes input data (e.g., images, text).
2. **Hidden Layers**: Perform computations and feature extraction. These layers can be fully connected (dense), convolutional, recurrent, etc.
3. **Output Layer**: Produces the final output (e.g., class probabilities, regression values).

Here's a simple diagram of a feedforward neural network:

```
Input Layer -> Hidden Layer 1 -> Hidden Layer 2 -> Output Layer
```

### Example Code

We'll use TensorFlow and Keras to build and train a simple MLP.

````python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Preprocess the data
X_train = X_train.reshape(-1, 28*28) / 255.0
X_test = X_test.reshape(-1, 28*28) / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the MLP
model = Sequential([
    Dense(128, activation='relu', input_shape=(28*28,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2)

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc:.4f}')
````

## 3. Example: MNIST Digit Classification <a id='example'></a>

In this example, we'll build and train a Multilayer Perceptron (MLP) to classify handwritten digits from the MNIST dataset. The dataset consists of 28x28 grayscale images of digits from 0 to 9.

````python
# Plotting the training history
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1])
plt.legend(loc='lower right')
plt.show()
````

##Other neural network models <a id='exercise'></a>

Do some research and choose a different neural network to train and repeat the previous exercise.

### Exercise Instructions:

1. **Dataset**: Fashion MNIST dataset from `tf.keras.datasets`
2. **Task**: Train a different neural network to classify fashion items.
3. **Steps**:
    - Load the Fashion MNIST dataset.
    - Preprocess the data.
    - Build a neural network different form the MLP.
    - Compile the model.
    - Train the model.
    - Evaluate the model on the test set.
    - Plot the training history.

````python
import tensorflow.keras.layers as layers

# Load the Fashion MNIST dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()

# Normalize the images to [0, 1]
train_images = train_images / 255.0
test_images = test_images / 255.0

# Reshape the images to include a channel dimension
train_images = train_images.reshape((train_images.shape[0], 28, 28, 1))
test_images = test_images.reshape((test_images.shape[0], 28, 28, 1))

# Build the CNN
model_CNN = Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model_CNN.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history_fm = model_CNN.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))
# Evaluate the model
test_loss_fm, test_acc_fm = model_CNN.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc_fm:.4f}')

# Plotting the training history
plt.plot(history_fm.history['accuracy'], label='accuracy')
plt.plot(history_fm.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1])
plt.legend(loc='lower right')
plt.show()
````
