```python
from PIL import Image
import numpy as np
import pathlib
import cv2
import matplotlib.pyplot as plt

# Cargar imagen del perro
dog = Image.open(pathlib.Path('animals/dog/00013.jpg'))
dog = np.array(dog) / 255.0

# Convertir la imagen a formato OpenCV
dog_cv = cv2.cvtColor((dog * 255).astype(np.uint8), cv2.COLOR_RGB2BGR)

# Ajustar contraste y brillo
dog_cv = cv2.convertScaleAbs(dog_cv, alpha=1.2, beta=30)

# Convertir a tonos azules: mantener el canal azul, reducir rojo y verde
dog_cv[:, :, 2] = dog_cv[:, :, 2] * 0.1  # Reducir el canal azul
dog_cv[:, :, 0] = dog_cv[:, :, 0] * 0.1  # Reducir el canal rojo

# Convertir de nuevo a formato PIL para mostrar
dog_green = Image.fromarray(cv2.cvtColor(dog_cv, cv2.COLOR_BGR2RGB))

# Mostrar la imagen final
plt.imshow(dog_green)
plt.axis('off') # Ocultar los ejes
plt.show()
# dog_blue.show()
```
